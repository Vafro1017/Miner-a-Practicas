```python
import seaborn as sn
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
columnas.append("totalyearlycompensation")
columnas
```

```python
Caux= []
for i in aux.columns:
    if i not in columnas:
        Caux.append(i)
        
Caux
```

```python
scaler.fit(aux.drop(Caux,axis=1))
```

```python
aux.drop(Caux,axis=1).info()
```

```python
scaled_features = scaler.transform(aux.drop(Caux,axis=1))
```

```python
Nonum = list(aux.drop(Caux,axis=1).columns)
Nonum
```

```python
df_feat = pd.DataFrame(scaled_features,columns=Nonum)
df_feat.head()
```

```python
sn.pairplot(data=df_feat)
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features,aux["gender"],test_size=0.30)
```

```python
aux['gender'].nunique()
```

```python
aux.dropna(subset=['gender'], inplace=True)
```

```python
aux.info()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
import numpy as np
```

```python
error_rate = []

# Will take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.pyplot.figure(figsize=(10,6))
plt.pyplot.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.pyplot.title('Error Rate vs. K Value')
plt.pyplot.xlabel('K')
plt.pyplot.ylabel('Error Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=1')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
